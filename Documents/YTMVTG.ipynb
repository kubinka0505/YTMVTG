```
#@title # <font color=red>⬅</font> **2.** Fill the data form (<a href="https://github.com/kubinka0505/YTMVTG/wiki/Configuration-Documentation">`Config.json`</a>) <font color=orange>**AND THEN**</font> press ▶
#@markdown ### `Aspect_Ratio`
Width = 16 #@param {type:"number"}
Height = 9 #@param {type:"number"}
#@markdown ### `URL`
Direct_URL = "https://avatars.githubusercontent.com/u/30776937" #@param {type:"string"}
#@markdown ### `Output`
Format = "PNG" #@param {type:"string"}
#@markdown ---  
Resizing_Algorithm = "3 (Linear / Bilinear)" #@param ["1 (Nearest / None)", "2 (Box)", "3 (Linear / Bilinear)", "4 (Hamming)", "5 (Cubic / Bicubic)", "6 (Antialias / Lanczos)"]
Resizing_Algorithm = int(Resizing_Algorithm.split("(")[0].strip())
```

```
#@title # <font color=red>⬅</font> **3.** Process and display `Image`
"""YouTube Music Videos Thumbnail Maker"""

import os
from PIL import Image
from time import time
from json import loads
from requests import get
from base64 import b64encode
from IPython.display import Image as display
os.chdir("/content/YTMVTG")

#--------------------------------#

__author__		= "kubinka0505"
__copyright__	= __author__
__credits__		= __author__
__version__		= "1.0"
__date__		= "16.03.2021"
__status__		= "Mature"
__license__		= "GPL V3"

#--------------------------------#

Config = loads(open("Config.json", encoding = "utf-8").read())
File_Path = Direct_URL

#--------------------------------#

Picture = Image.open(get(File_Path, stream = True).raw)

if not Format:
	Format = "PNG"

Name = "YTMLTM_" + str(time()).split(".")[0]
Format = Format.lower()

Width, Height = Picture.size[0] * Config["Aspect_Ratio"]["Width"] // Config["Aspect_Ratio"]["Height"], Picture.size[1]

Background_Color = Picture.resize((1,1), Config["Resizing_Algorithm"]).getpixel((0,0))
Background = Image.new(size = (Width, Height), mode = Picture.mode, color = Background_Color)
Background.paste(Picture, ((Background.size[0] - Picture.size[0]) // 2, (Background.size[1] - Picture.size[1]) // 2))
Background.save("Images/{0}.{1}".format(Name, Format), quality = 100)

print("Displaying... (This can take a while)\n")

display("/content/YTMVTG/Images/{0}.{1}".format(Name, Format), width = 500)
```
