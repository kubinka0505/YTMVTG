```
#@title # <font color=red>⬅</font> **2.** Fill the data form (<a href="https://github.com/kubinka0505/YTMVTG/wiki/Configuration-Documentation">`Config.json`</a>) <font color=orange>**AND THEN**</font> press ▶
#@markdown ### `Target_Aspect_Ratio`
Width = 16 #@param {type:"number"}
Height = 9 #@param {type:"number"}
#@markdown ---  
Direct_URL = "https://avatars.githubusercontent.com/u/30776937" #@param {type:"string"}
Output_Format = "PNG" #@param {type:"string"}
Resizing_Algorithm = "3 (Linear / Bilinear)" #@param ["1 (Nearest / None)", "2 (Box)", "3 (Linear / Bilinear)", "4 (Hamming)", "5 (Cubic / Bicubic)", "6 (Antialias / Lanczos)"]
Resizing_Algorithm = int(Resizing_Algorithm.split("(")[0].strip())
Background_Brightness_Percentage = 100 #@param {type:"number"}
Mode = "Resizing Algorithms Comparison" #@param ["Default", "Resizing Algorithms Comparison"]
```

```
#@title # <font color=red>⬅</font> **3.** Process and display `Image`
"""YouTube Music Videos Thumbnail Maker"""

import os
from json import loads
from requests import get
from time import time, sleep
from PIL import Image, ImageEnhance
from IPython.display import display, Image as Image_
os.chdir("/content/YTMVTG")

#--------------------------------#

__author__		= "kubinka0505"
__copyright__	= __author__
__credits__		= __author__
__version__		= "1.2"
__date__		= "18.03.2021"
__status__		= "Mature"
__license__		= "GPL V3"

#--------------------------------#

Config = loads(open("Config.json", encoding = "utf-8").read())
File_Path = Direct_URL
Filters = [Image.NONE, Image.BOX, Image.LINEAR, Image.HAMMING, Image.CUBIC, Image.LANCZOS]

#--------------------------------#

Picture = Image.open(get(File_Path, stream = True).raw)

if not Output_Format:
	Output_Format = "PNG"

Name = "YTMLTM_" + str(time()).split(".")[0]
Output_Format = Output_Format.lower()

Width_, Height_ = Picture.size[0] * Width // Height, Picture.size[1]

print("Displaying... (This can take a while)")
if Mode == "Default":
	Background_Color = Picture.resize((1,1), Resizing_Algorithm)
	Background_Color = ImageEnhance.Brightness(Background_Color).enhance(Background_Brightness_Percentage / 200)
	Background_Color = Background_Color.getpixel((0,0))
	#---#
	Background = Image.new(Picture.mode, (Width_, Height_), Background_Color)
	Background.paste(Picture, ((Background.size[0] - Picture.size[0]) // 2, (Background.size[1] - Picture.size[1]) // 2))
	#---#
	Background.save("Images/{0}.{1}".format(Name, Output_Format), quality = 100)
	display(Image_("/content/YTMVTG/Images/{0}.{1}".format(Name, Output_Format), width = 500))
else:
	Background_Comparison = Image.new(Picture.mode, (Width_, Height_ * len(Filters)))
	Y = 0
	#---#
	for Algorithm in Filters:
		Background_Color = Picture.resize((1,1), Algorithm)
		Background_Color = ImageEnhance.Brightness(Background_Color).enhance(Background_Brightness_Percentage / 200)
		Background_Color = Background_Color.getpixel((0,0))
		#---#
		Background = Image.new(Picture.mode, (Width_, Height_), Background_Color)
		Background.paste(Picture, ((Background.size[0] - Picture.size[0]) // 2, (Background.size[1] - Picture.size[1]) // 2))
		#---#
		Background_Comparison.paste(Background, (0, Y))
		Y += Height_
	#---#
	Background_Comparison.save("/tmp/YTMVTG.png")
	display(Image_("/tmp/YTMVTG.png", width = 250))
 	#---#
	os.remove("/tmp/YTMVTG.png")
```
